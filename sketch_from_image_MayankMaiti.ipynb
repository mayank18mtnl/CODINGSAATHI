# **Image to pencil sketch using python**
# LEVEL 1 Task 4
To conver a image into pencil sketch with the help of python (OpenCV)

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/images/5mage5.jpeg")
cv2_imshow(image)
```

```python
gr_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gr_img)
```

```python
ivt= cv2.bitwise_not(gr_img)
cv2_imshow(ivt)
```

```python
blur=cv2.GaussianBlur(ivt,(21,21),0)
cv2_imshow(blur)
```

```python
ivt_blur = cv2.bitwise_not(blur)
cv2_imshow(ivt_blur)
```

```python
sketch = cv2.divide(gr_img,ivt_blur,scale=255.0)
cv2_imshow(sketch)
cv2.imwrite("sketch1.jpg",sketch)
```
