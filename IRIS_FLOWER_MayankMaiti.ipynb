```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columns =['Sepal length','sepal width','petal length','petal wdth','class-labels']
df=pd.read_csv('/iris.data',names=columns)
df.head(150)
```

```python
df.describe()
```

```python
sns.pairplot(df,hue='class-labels')
```

```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train,y_train)
```

```python
outcome1 = model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print("Accuracy Score = ",accuracy_score(y_test,outcome1)*100)
```

```python
for i in range(len(outcome1)):
  print(y_test[i],outcome1[i])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,outcome1))
```
